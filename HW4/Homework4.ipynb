```python
import pandas as pd
df_phones = pd.read_csv('cell_phones_total.csv')
```

Before I start, I will replicate  the code from the previous homework in order to clean the data. 

```python

# Conversion function
def convert_to_numeric(value):
    if pd.isnull(value):
        return value  # Preserve NaN for filling later
    elif value == '0':
        return 0
    else:
        try:
            return float(value)
        except ValueError:
            multiplier = 1
            if 'K' in value:
                multiplier = 1e3
            elif 'M' in value:
                multiplier = 1e6
            elif 'B' in value:
                multiplier = 1e9

            # Handle scientific notation
            if 'e' in value.lower():
                return float(value)

            # Extract numeric part and multiply
            numeric_part = float(''.join(filter(str.isdigit, value)))
            return numeric_part * multiplier

# Apply the conversion function and handle missing values
columns_to_convert = ['1960', '1965', '1966', '1967', '1968', '1969', '1970', '1971', 
                      '1972', '1973', '1974', '1975', '1976', '1977', '1978', '1979','1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989',
    '1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999',
    '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009',
    '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']

# Apply numeric conversion
for column in columns_to_convert:
    df_phones[column] = df_phones[column].apply(convert_to_numeric)

df_phones[columns_to_convert] = df_phones[columns_to_convert].ffill(axis=1)  # Forward fill
df_phones[columns_to_convert] = df_phones[columns_to_convert].bfill(axis=1)  # Backward fill

pd.set_option('display.max_columns', None)
df_phones.head(10)
```

Let's view the population data! this data set contains data about the population size of countries over years. The missing values will, just like with the cell phone data, use forward fill and backward fill in order to deal with null values. The last step prints out the information after the cleaned data frame

```python
df_pop = pd.read_csv('pop_data.csv')


df_pop = df_pop.fillna(pd.NA)
# Forward fill missing values
df_pop_ffill = df_pop.ffill()

# Backfill missing values
df_pop_bfill = df_pop.bfill()

print("\nDataset Info after Filling Missing Values with back fill/forward fill:")
print(df_pop.info())
df_pop
```

Next step is to combine the cell phone data with the population data. This is to be done by pivotting the tables into the right format using the melt and merge functions. 

```python
df_phones_melted = pd.melt(df_phones, id_vars=['iso-3'], var_name='year', value_name='n_cellphones')

df_pop_melted = pd.melt(df_pop, id_vars=['iso-3'], var_name='year', value_name='population')

result_df = pd.merge(df_phones_melted, df_pop_melted, on=['iso-3', 'year'])

result_df.index
result_df.head(10)
```

Country_data is used to map the iso-3 code to the country name with iso-3 as an intermediate. This is then pivoted into a table with year as an index, country names as columns and phones per capita as values.

I will now combine the cell phone, population and country data into a single table by mapping between iso-3 and country names.

```python
# Step 1: Read the country_data.csv
country_data = pd.read_csv('country_data.csv')

# Step 2: Merge the result_df  with country_data 
merged_df = pd.merge(result_df, country_data, left_on='iso-3', right_on='alpha-3')

# Step 3: Drop unnecessary columns
merged_df = merged_df.drop(columns=['iso-3', 'alpha-3'])

# The resulting DataFrame
result_df.head(10)
```

Now I will create the new column that show cell phones per capita by using the data of the number of cell phones and the population. Next I will make us of a pivot table to get year as the row index, country names as column heads and cells as the calculated phones per capita values.

```python
merged_df['phones_per_capita'] = merged_df['n_cellphones'] / merged_df['population']

phones_per_capita_df = merged_df.pivot(index='year', columns='name', values='phones_per_capita')

phones_per_capita_df


```

As asked by the specific task, I will present a table with countries that had the highest number of cell phones per capita in 2019.

```python
phones_per_capita_df.index = phones_per_capita_df.index.astype(int)

phones_2019 = phones_per_capita_df.loc[2019]

top_5_countries_2019 = phones_2019.nlargest(5).index

filtered_phones_per_capita_df = phones_per_capita_df.loc[2015:2019, top_5_countries_2019]

filtered_phones_per_capita_df

```

# SQL

First task is to retrieve the usernames of all users who have performed the "signup" action.

```python
import sqlite3

# Connect to the SQLite database
conn = sqlite3.connect('user_actions.db')
cursor = conn.cursor()

# Task 1: Retrieve the usernames of all users who have performed the "signup" action.
cursor.execute("SELECT DISTINCT username FROM user_actions WHERE action = 'signup'")
result_task1 = cursor.fetchall()

print("Task 1 Result:")
for row in result_task1:
    print(row[0])
```

Next task is to find the total number of log entries for each user. Display the user_id, username, and the count of log entries.

```python
# Task 2: Find the total number of log entries for each user. Display the user_id, username, and the count of log entries.
cursor.execute("SELECT user_id, username, COUNT(*) AS log_entry_count FROM user_actions GROUP BY user_id, username")
result_task2 = cursor.fetchall()

print("\nTask 2 Result:")
for row in result_task2:
    print(f"User ID: {row[0]}, Username: {row[1]}, Log Entry Count: {row[2]}")
```

Next is to identify users who have both logged in (action = 'login') and signed up (action = 'signup') on the same day while displaying the user_id and username.

```python
# Task 3: Identify users who have both logged in (action = 'login') and signed up (action = 'signup') on the same day. Display the user_id and username.
query_task3 = """
    SELECT ua1.user_id, ua1.username
    FROM user_actions ua1
    JOIN user_actions ua2 ON ua1.user_id = ua2.user_id
    WHERE ua1.action = 'login'
      AND ua2.action = 'signup'
      AND DATE(ua1.timestamp) = DATE(ua2.timestamp);
"""

cursor.execute(query_task3)
result_task3 = cursor.fetchall()

print("\nTask 3 Result:")
for row in result_task3:
    print(f"User ID: {row[0]}, Username: {row[1]}")

# Close the connection
conn.close()
```

# Regex

The first task is to write a regular expression to extract all hashtags from a given comment.

```python
import re

with open('comments.txt', 'r') as file:
    comments = file.read()
def extract_comment_by_number(comment_number, comments_text):
    # Regular expression to match a comment with a specific number
    regex = fr"{comment_number}\. \"(.*?)\""

    # Use re.search to find the matched comment
    match = re.search(regex, comments_text)

    # If a match is found, return the comment text; otherwise, return None
    if match:
        return match.group(1)
    else:
        return None
comment_1 = extract_comment_by_number(1,comments)
# Regular expression to extract hashtags
hashtags_regex = r"#\w+"

# Extract hashtags from the comment
hashtags_comment1 = re.findall(hashtags_regex, comment_1)
print(hashtags_comment1)
```

Next task is to create a regular expression to find comments that mention both "#programming" and "#python".

```python
regex_both_tags = r"(?=.*#programming)(?=.*#python)"

comment_2 = extract_comment_by_number(2,comments)
# Check if the regex matches the comment
match_comment2 = re.match(regex_both_tags, comment_2)

print(bool(match_comment2))
```
